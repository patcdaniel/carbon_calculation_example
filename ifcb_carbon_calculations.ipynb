```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('habdac_query_output_conc.csv')
proro_conc = df['Dinophysis spp.']
time = pd.to_datetime(df['timestamp'])
carbon = 0.216 * (proro_conc)**.939
```

```python
fig, ax = plt.subplots()

x = np.linspace(0, 10e7, 1000)
ax.scatter(proro_conc, carbon)

ax.set_xscale('log')
ax.set_yscale('log')
```

```python
fig, (ax,ax2) = plt.subplots(2)
fig.set_size_inches(10, 7)
ax.bar(time, proro_conc,width=0.035)
ax.set_ylim(0,60000)
ax.set_ylabel("Biovolume (um^3/ml)")

ax2.bar(time, carbon,width=0.035)
ax2.set_ylabel("Carbon (ugC/L)")
fig.autofmt_xdate()
# ax2.set_ylim(0,5500)
```
